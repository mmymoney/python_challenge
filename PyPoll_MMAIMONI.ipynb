```python
#import csv and os

import csv
import os
import statistics
import numpy as np
```

```python
#define file variable for imported csv
file = 'PyPoll/Resources/election_data.csv'
```

```python
#Voting Calculations

with open(file, newline="") as csvfile_b:
    #Store header row
    header = next(csvfile_b)
    #variable for instances of candidates
    candidates = []
    #for loop to collect candidates and add to list
    for x in csv.reader(csvfile_b):
            candidates.append(x[2])
            
    #Aggregate votes for total vote count
    total_votes = len(candidates)
    
    #create variables for each candidate's votes
    khan = []
    correy = []
    li = []
    otooley = []
    
    #define function to count number of instances for each candidate
    def candidate_ct(a,b):
        return a.count(b)
    
    #Khan vote total
    a = candidates
    b = "Khan"
    
    khan = (b, candidate_ct(a,b))
    
    #Correy vote total
    a = candidates
    b = "Correy"
    
    correy = (b, candidate_ct(a,b))
  
    #Li vote total
    a = candidates
    b = "Li"

    li = (b, candidate_ct(a,b))

    #O'Tooley vote total
    a = candidates
    b = "O'Tooley"

    otooley = (b, candidate_ct(a,b))

    #Calculate percent of total votes each candidate received
    khanpct = "{:.3%}".format(khan[1]/total_votes)
    correypct = "{:.3%}".format(correy[1]/total_votes)
    lipct = "{:.3%}".format(li[1]/total_votes)
    otooleypct = "{:.3%}".format(otooley[1]/total_votes)
    
    #Winner Selection

    winnervotes = max(khan[1], correy[1], li[1], otooley[1])
    winnerid = ""
    if int(winnervotes) == int(khan[1]):
        winnerid = "Khan"
    elif int(winnervotes) == int(correy[1]):
        winnerid="Correy"
    elif int(winnervotes) == int(li[1]):
        winnerid ="Li"
    elif int(winnervotes) == int(otooley[1]):
        winnerid = "O'Tooley"
    else:
        winnerid = "No Identified Winner"
        
    
    #Export to Text File and Print to Screen
    print(f'PyPoll')
    print(f'Election Results')
    print(f'--------------------')
    print(f'Total Votes: {str(total_votes)}')
    print(f'--------------------')
    print(f'Khan: {khanpct} ({khan[1]})')
    print(f'Correy: {correypct} ({correy[1]})')
    print(f'Li: {lipct} ({li[1]})')
    print(f"O'Tooley: {otooleypct} ({otooley[1]})")
    print(f'--------------------')
    print(f'Winner: {winnerid}')
    print(f'--------------------')

    
    #Let's export to text...
    file_export = open("PyPoll Analysis","w")
    file_export.writelines(f'PyPoll \n')
    file_export.writelines(f'Election Results \n')
    file_export.writelines(f'-------------------- \n')
    file_export.writelines(f'Total Votes: {str(total_votes)} \n')
    file_export.writelines(f'-------------------- \n')
    file_export.writelines(f'Khan: {khanpct} ({khan[1]}) \n')
    file_export.writelines(f'Correy: {correypct} ({correy[1]}) \n')
    file_export.writelines(f'Li: {lipct} ({li[1]}) \n')
    file_export.writelines(f"O'Tooley: {otooleypct} ({otooley[1]} \n")
    file_export.writelines(f'-------------------- \n')
    file_export.writelines(f'Winner: {winnerid} \n')
    file_export.writelines(f'--------------------')
    file_export.close    
```
